```python
import pandas as pd
import tensorflow
import sklearn
```

```python
from tensorflow.keras.layers import Dense, Dropout, Activation, Input
from tensorflow.keras.models import Model
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
base=base = pd.read_csv('games.csv')
```

```python
base = base.drop('Other_Sales', axis = 1)
base = base.drop('Global_Sales', axis = 1)
base = base.drop('Developer', axis = 1)
```

```python
base = base.dropna(axis = 0)
```

```python
base = base.drop('Name', axis = 1)
```

```python
X = base.iloc[:, [0, 1, 2, 3, 7, 8, 9, 10, 11]].values
```

```python
y_na = base.iloc[:, 4].values
y_eu = base.iloc[:, 5].values
y_jp = base.iloc[:, 6].values
y_ot = base.iloc[:, 7].values
y_gl = base.iloc[:, 8].values
```

```python
onehotencoder = ColumnTransformer(transformers=[("OneHot", OneHotEncoder(), [0, 2, 3, 8])], remainder='passthrough')
X = onehotencoder.fit_transform(X).toarray()
```

```python
X.shape
```

```python
camada_entrada = Input(shape = (303,))
camada_oculta1 = Dense(units = 153, activation='relu')(camada_entrada)
camada_oculta2 = Dense(units = 153, activation='relu')(camada_oculta1)
camada_oculta2 = Dense(units = 153, activation='relu')(camada_oculta1)
camada_saida1 = Dense(units = 1, activation='linear')(camada_oculta2)
camada_saida2 = Dense(units = 1, activation='linear')(camada_oculta2)
camada_saida3 = Dense(units = 1, activation='linear')(camada_oculta2)
camada_saida4 = Dense(units = 1, activation='linear')(camada_oculta2)
camada_saida5 = Dense(units = 1, activation='linear')(camada_oculta2)
```

```python
regressor = Model(inputs = camada_entrada, outputs = [camada_saida1, camada_saida2, camada_saida3, camada_saida4,camada_saida5])
```

```python
regressor.compile(optimizer='adam', loss = 'mse')
```

```python
regressor.fit(X, [y_na, y_eu, y_jp, y_ot, y_gl], epochs=500, batch_size=100)
```

```python
previsao_na, previsao_eu, previsao_jp, previsao_ot, previsao_gl= regressor.predict(X)
```

```python
previsao_na, previsao_na.mean(), previsao_eu, previsao_eu.mean(),previsao_jp, previsao_jp.mean(), previsao_ot, previsao_ot.mean(), previsao_gl, previsao_gl.mean()
 
```
