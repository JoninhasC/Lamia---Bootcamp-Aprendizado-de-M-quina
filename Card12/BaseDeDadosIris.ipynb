```python
import pandas as pd
import numpy as np
import tensorflow as tf 
import sklearn
from tensorflow.python.keras.utils import np_utils
```

```python
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import accuracy_score
```

```python
base = pd.read_csv('iris.csv')
```

```python
base
```

```python
# Seleciona as colunas de índice 0 a 3 do DataFrame 'base' e armazena os valores em 'x'
X = base.iloc[:, 0:4].values  # A variável 'x' agora contém os dados das quatro primeiras colunas da base, usadas como características de entrada
```

```python
X
```

```python
# Seleciona a coluna de índice 4 do DataFrame 'base' e armazena os valores em 'y'
y = base.iloc[:, 4].values  # A variável 'y' agora contém os dados da quinta coluna da base, que é usada como rótulo para o modelo
```

```python
y
```

```python
# Cria uma instância do codificador de rótulos para transformar rótulos de texto em valores numéricos
labelencoder = LabelEncoder()

# Aplica o codificador ao vetor 'y' para converter as classes de rótulos em valores inteiros
y = labelencoder.fit_transform(y)  # Cada rótulo agora é representado por um número único
```

```python
y
```

```python
# Converte a variável `y` para uma matriz categórica (one-hot encoding)
y = to_categorical(y)  # Define num_classes se necessário

y  # Mostra o valor transformado
```

```python
# Divide o conjunto de dados em dados de treinamento e teste
# x_train e y_train são usados para treinar o modelo
# x_test e y_test são usados para avaliar a performance do modelo
# test_size=0.25 indica que 25% dos dados serão reservados para teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Cria uma nova rede neural sequencial
rede_neural = Sequential([
    # Camada de entrada: recebe dados com 4 características
    tf.keras.layers.InputLayer(shape=(4,)),
    
    # Primeira camada oculta: camada densa com 4 neurônios e função de ativação ReLU
    tf.keras.layers.Dense(units=4, activation='relu'),
    
    # Segunda camada oculta: outra camada densa com 4 neurônios e função de ativação ReLU
    tf.keras.layers.Dense(units=4, activation='relu'),
    
    # Camada de saída: camada densa com 3 neurônios para previsão de classes múltiplas
    # A função de ativação softmax é utilizada para normalizar as saídas entre 0 e 1
    tf.keras.layers.Dense(units=3, activation='softmax')
])
```

```python
rede_neural.summary()
```

```python
 #Compila o modelo da rede neural
# Define o otimizador como 'adam', a função de perda como 'categorical_crossentropy' para problemas de classificação multiclasse,
# e a métrica como 'categorical_accuracy' para avaliar a acurácia do modelo durante o treinamento
rede_neural.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['categorical_accuracy'])
```

```python
# Treina a rede neural com os dados de treinamento
# X_train e y_train são as variáveis que contêm os dados de entrada e as classes correspondentes
rede_neural.fit(X_train, y_train, batch_size=10, epochs=1500)  # Tamanho do lote de 10 e 1500 épocas
```

```python
rede_neural.evaluate(X_test, y_test)
```

```python
previsoes = rede_neural.predict(X_test)
print(previsoes)
```

```python
previsoes = previsoes > 0.5
print(previsoes)
```
