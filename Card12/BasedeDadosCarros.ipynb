```python
# Importa a classe Sequential para criar um modelo sequencial, que organiza as camadas em uma sequência linear.
regressor = Sequential([
    
    # Define a camada de entrada usando InputLayer.
    # 'shape=(316,)' indica que o modelo espera entradas com 316 características (ou variáveis independentes).
    tf.keras.layers.InputLayer(shape=(316,)),
    
    # Primeira camada densa (totalmente conectada) com 158 unidades (neurônios).
    # 'activation='relu'' define a função de ativação ReLU (Rectified Linear Unit),
    # que introduz não-linearidade no modelo, permitindo que ele aprenda representações mais complexas.
    tf.keras.layers.Dense(units=158, activation='relu'),
    
    # Segunda camada densa com 158 unidades e função de ativação ReLU.
    # Camadas adicionais ajudam a rede a aprender padrões complexos ao aumentar a profundidade do modelo.
    tf.keras.layers.Dense(units=158, activation='relu'),
    
    # Camada de saída com apenas 1 unidade, pois estamos prevendo um valor contínuo.
    # A função de ativação 'linear' é utilizada aqui, apropriada para regressão,
    # pois a saída é um valor contínuo (sem limitação entre 0 e 1, ou -1 e 1).
    tf.keras.layers.Dense(units=1, activation='linear')
])
```

```python
regressor.summary()
```

```python
# Compilar o modelo com a função de perda correta
regressor.compile(optimizer='adam', loss='mean_absolute_error', metrics=['mean_absolute_error'])
```

```python
# Inicia o processo de treinamento do modelo 'regressor' usando o método fit(), que ajusta o modelo aos dados fornecidos.

regressor.fit(
    X,  # 'X' representa os dados de entrada (características ou variáveis independentes) usados para o treinamento.
    y,  # 'y' é o rótulo ou variável de saída (valor alvo) que o modelo deve aprender a prever.
    
    # 'batch_size=300' define o número de amostras de treinamento processadas antes de atualizar os pesos do modelo.
    # Nesse caso, o modelo processa 300 amostras antes de cada atualização dos parâmetros internos.
    # Um valor de batch size maior pode acelerar o treinamento, mas exige mais memória.
    
    batch_size=300,
    
    # 'epochs=100' define o número de vezes que o modelo verá o conjunto completo de dados durante o treinamento.
    # Treinar por mais épocas permite que o modelo ajuste melhor seus parâmetros, porém pode aumentar o risco de overfitting.
    # Neste caso, o modelo será treinado em 100 iterações sobre o conjunto de dados.
    
    epochs=100
)
```

```python
def criar_rede():
    # Limpa a sessão atual do backend do TensorFlow para evitar o acúmulo de modelos em memória.
    # Isso é útil para garantir que uma nova rede seja criada do zero, sem interferência de redes anteriores.
    tf.keras.backend.clear_session()
    
    # Inicializa o modelo sequencial, que é uma pilha linear de camadas.
    regressor = Sequential([
        
        # Define a camada de entrada com 316 neurônios de entrada.
        # Isso indica que o modelo espera receber dados de entrada com 316 características.
        tf.keras.layers.InputLayer(shape=(316,)),
        
        # Primeira camada densa com 158 neurônios e função de ativação ReLU.
        # A função de ativação ReLU introduz não-linearidade, permitindo que o modelo capture relações complexas nos dados.
        tf.keras.layers.Dense(units=158, activation='relu'),
        
        # Segunda camada densa com 158 neurônios e função de ativação ReLU.
        # Esta camada adicional aprofunda a rede, aumentando sua capacidade de aprendizado.
        tf.keras.layers.Dense(units=158, activation='relu'),
        
        # Camada de saída com 1 neurônio e ativação linear.
        # A ativação linear é apropriada para regressão, pois permite uma saída contínua não limitada.
        tf.keras.layers.Dense(units=1, activation='linear'),
    ])
    
    # Compila o modelo definindo o otimizador, a função de perda e as métricas.
    # 'optimizer='adam'' define o otimizador Adam, um dos mais eficientes para redes neurais,
    # que ajusta os pesos do modelo durante o treinamento.
    # 'loss='mean_absolute_error'' define a função de perda como erro absoluto médio (MAE),
    # uma métrica comum para tarefas de regressão que calcula a média dos erros absolutos.
    # 'metrics=['mean_absolute_error']' especifica que o MAE também será utilizado como métrica
    # durante o treinamento para monitorar o desempenho do modelo.
    regressor.compile(optimizer='adam', loss='mean_absolute_error', metrics=['mean_absolute_error'])
    
    # Retorna o modelo compilado, pronto para ser treinado com os dados.
    return regressor
```
